```python
import pandas as pd
df1 = pd.read_csv('data/DENSE_SGD_2022-09-30_23_09.csv')
df1['model_type'] = 'SGD Optimizer'
df2 = pd.read_csv('data/DENSE_Adam_2022-09-30_22_12.csv')
df2['model_type'] = 'Adam Optimizer'
df3 = pd.read_csv('data/DENSE_Adam_Dropout_2022-09-30_22_34.csv')
df3['model_type'] = 'Adam with Dropout'
df = pd.concat([df1,df2,df3],ignore_index=True)
df.to_csv('data/ALL_DENSE_MODELS.csv',index=False)
display(df)
df[['elapsed_time','test_accuracy']].describe()
```

```python
## What are the minimum and maximum accuracy scores across all 36 models?
ans = df.describe()
min_score = ans.loc['min','test_accuracy']
print(f'{min_score=}')
max_score = ans.loc['max','test_accuracy']
print(f'{max_score=}')
```

```python
## What are the minimum and maximum training times in seconds across all 36 models?
min_time = ans.loc['min','elapsed_time']
max_time = ans.loc['max','elapsed_time']
print(f'{min_time=}')
print(f'{max_time=}')
```

```python
## Which model configurations had the longest and shortest training times?
min_row = df.iloc[df['elapsed_time'].argmin(),:]
display(pd.DataFrame(min_row).T)
max_row = df.iloc[df['elapsed_time'].argmax(),:]
display(pd.DataFrame(max_row).T)

```

```python
## Which model configuratons had the best and worst test accuracy?
min_row = df.iloc[df['test_accuracy'].argmin(),:]
display(pd.DataFrame(min_row).T)
max_row = df.iloc[df['test_accuracy'].argmax(),:]
display(pd.DataFrame(max_row).T)
```
